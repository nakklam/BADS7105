# Prepare data

```python
# Investigate all question
import re

# cleansing column name
# interest_column = []
# consume_column = []
for col in cust_df.columns:
    if col.startswith("คุณมีความสนใจ"):
        text_in_brackket = re.match(r"[^[]*\[([^]]*)\]", col).groups()[0]
        new_col_name = 'สนใจ '+text_in_brackket
        cust_df.rename(columns={col:new_col_name},inplace=True)
#         interest_column.append(new_col_name)
        
    elif col.startswith("คุณบริโภค"):
        text_in_brackket = re.match(r"[^[]*\[([^]]*)\]", col).groups()[0]
        new_col_name = 'บริโภค '+text_in_brackket
        cust_df.rename(columns={col:new_col_name},inplace=True)
        print(new_col_name)
#         consume_column.append(new_col_name)

for col in cust_df.columns:
    print(col)
    
# print(interest_column)
# print(consume_column)
```

```python
# Convert to number
interest_choice_dict = {
    'ไม่สนใจอย่างมากที่สุด' : 0,
    'ไม่สนใจอย่างมาก' : 1,
    'ไม่สนใจ' : 2,
    'เฉยๆ' : 3,
    'สนใจ' : 4,
    'สนใจอย่างมาก' : 5,
    'สนใจอย่างมากที่สุด' : 6,
}

consume_choice_dict = {
    'แทบไม่ได้บริโภคเลย': 0,
    'หลายเดือนครั้ง' : 1,
    'เดือนละ 2-3 ครั้ง' : 2,
    'เดือนละครั้ง' : 3,
    'อาทิตย์ละครั้ง' : 4,
    'แทบทุกวัน' : 5,
}

interest_column = [col for col in cust_df.columns if col.startswith('สนใจ')]
consume_column = [col for col in cust_df.columns if col.startswith('บริโภค')]

# encode to be number
cust_df[interest_column] = cust_df[interest_column].replace(interest_choice_dict)
cust_df[consume_column] = cust_df[consume_column].replace(consume_choice_dict)
cust_df.head()
```

```python
# แก้ไขข้อมูล พศ ให้เป็น คศ

from datetime import datetime, date 
from dateutil.relativedelta import relativedelta

d1 = datetime(2021, 3, 1)
# แปลงอายุที่กรอกเป็นปี พส.
cust_df.loc[cust_df['คุณเกิดวันที่เท่าไหร่']>d1, 'คุณเกิดวันที่เท่าไหร่'] = cust_df['คุณเกิดวันที่เท่าไหร่']-relativedelta(years=543)
cust_df.head()
```

```python
# calculate age 
def calculate_age(born):
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

cust_df['อายุ'] = cust_df['คุณเกิดวันที่เท่าไหร่'].apply(calculate_age)
# cust_df.drop(['คุณเกิดวันที่เท่าไหร่'],axis=1,inplace=True)
cust_df['กลุ่มอายุ'] = pd.cut(cust_df['อายุ'], bins=[0,28,32,36,99], right=False, labels=['<28','28-31', '32-35','>36'])
cust_df.head()
```

# Analytic by using Pearson Correlation

```python
# หาคู่ขอข้อมูลที่มีค่า Correlation สัมพันธ์กันมากสุด
def get_redundant_pairs(df):
    '''Get diagonal and lower triangular pairs of correlation matrix'''
    pairs_to_drop = set()
    cols = df.columns
    for i in range(0, df.shape[1]):
        for j in range(0, i+1):
            pairs_to_drop.add((cols[i], cols[j]))
    return pairs_to_drop

def get_top_most_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=False)
    return au_corr[0:n]

def get_top_least_correlations(df, n=5):
    au_corr = df.corr().abs().unstack()
    labels_to_drop = get_redundant_pairs(df)
    au_corr = au_corr.drop(labels=labels_to_drop).sort_values(ascending=True)
    return au_corr[0:n]


print("Top 20 Most Correlations\n")
cor_df = cust_df[interest_column+consume_column]
print(get_top_most_correlations(cor_df, 20))
print("-------------------------\n")
print("Top 20 Least Correlations\n")
print(get_top_least_correlations(cor_df, 20))
```

```python
group_list = ['ชาย','หญิง']
for group in group_list:
    print("===========================")
    print("Top most of group : ",group)
    print("===========================")
    age_group_df = cust_df[cust_df['เพศ']==group]
    cor_df = age_group_df[interest_column+consume_column]
    print(get_top_most_correlations(cor_df, 5))
    print('\n')
```
