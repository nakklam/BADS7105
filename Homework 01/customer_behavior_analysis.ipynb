# Prepare data

```python
import re

# cleansing column name
for col in cust_df.columns:
    if col.startswith("คุณมีความสนใจ"):
        text_in_brackket = re.match(r"[^[]*\[([^]]*)\]", col).groups()[0]
        new_col_name = 'สนใจ '+text_in_brackket
        cust_df.rename(columns={col:new_col_name},inplace=True)
    elif col.startswith("คุณบริโภค"):
        text_in_brackket = re.match(r"[^[]*\[([^]]*)\]", col).groups()[0]
        new_col_name = 'บริโภค '+text_in_brackket
        cust_df.rename(columns={col:new_col_name},inplace=True)

for col in cust_df.columns:
    print(col)
```

```python
# แก้ไขข้อมูล พศ ให้เป็น คศ

from datetime import datetime, date 
from dateutil.relativedelta import relativedelta

d1 = datetime(2021, 3, 1)
# แปลงอายุที่กรอกเป็นปี พส.
cust_df.loc[cust_df['คุณเกิดวันที่เท่าไหร่']>d1, 'คุณเกิดวันที่เท่าไหร่'] = cust_df['คุณเกิดวันที่เท่าไหร่']-relativedelta(years=543)
cust_df.head(5)
```

```python
# calculate age 
def calculate_age(born):
#     born = datetime.strptime(born, "%d/%m/%Y").date()
    today = date.today()
    return today.year - born.year - ((today.month, today.day) < (born.month, born.day))

cust_df['อายุ'] = cust_df['คุณเกิดวันที่เท่าไหร่'].apply(calculate_age)
cust_df.head()
```

```python
# Conveert Likert Scale
column_interest = [col for col in cust_df.columns if col.startswith('สนใจ')]
column_consume = [col for col in cust_df.columns if col.startswith('บริโภค')]

print(column_interest)
print(column_consume)
```

```python
# convert questionare to number
interest_to_numberic = {
    'ไม่สนใจอย่างมากที่สุด' : 1,
    'ไม่สนใจอย่างมาก' : 2,
    'ไม่สนใจ' : 3,
    'เฉยๆ' : 4,
    'สนใจ' : 5,
    'สนใจอย่างมาก' : 6,
    'สนใจอย่างมากที่สุด' : 7
}

consume_to_numberic = {
    'แทบไม่ได้บริโภคเลย' : 1,
    'หลายเดือนครั้ง' : 2,
    'เดือนละครั้ง' : 3,
    'เดือนละ 2-3 ครั้ง' : 4,
    'อาทิตย์ละครั้ง' : 5,
    'แทบทุกวัน' : 6,
}


cust_df[column_interest] = cust_df[column_interest].replace(interest_to_numberic)
cust_df[column_consume] = cust_df[column_consume].replace(consume_to_numberic)
cust_df.head()
```
